```python
import random

def num_guess():
    print("Pilih tingkat kesulitan:")
    print("==" * 10)

    while True:
        level = input("1/2/3: ").strip()

        if level == "1":
            batas, max_coba = 50, 3
        elif level == "2":
            batas, max_coba = 100, 7
        elif level == "3":
            batas, max_coba = 200, 14
        else:
            print("Pilihan anda tidak valid")
            continue

        angka_rahasia = random.randint(1, batas)
        tebakanList = []

        print(f"\nTebak angka antara 1 sampai {batas}!")
        print(f"Kamu punya {max_coba} kesempatan.\n")

        percobaan = 1
        while percobaan <= max_coba:
            sisa = max_coba - percobaan + 1
            print(f"[Percobaan {percobaan}/{max_coba}] Sisa: {sisa} kesempatan")

            try:
                tebakan = int(input("Tebakan kamu: "))
            except ValueError:
                print(" Masukkan angka yang valid!\n")
                continue  # tidak menambah percobaan

            tebakanList.append(tebakan)

            if tebakan < angka_rahasia:
                print("Terlalu KECIL!\n")
            elif tebakan > angka_rahasia:
                print("Terlalu BESAR!\n")
            else:
                print(f"SELAMAT! Kamu berhasil menebak angka {angka_rahasia}!")
                print(f"Butuh {percobaan} percobaan.")
                print(f"Riwayat tebakan: {tebakanList}")
                break

            percobaan += 1
        else:
            print(f"\n GAME OVER! Angka yang benar adalah: {angka_rahasia}")
            print(f"   Riwayat tebakan: {tebakanList}")
            break

        ulang = input("\nMau main lagi? (y/n): ").strip().lower()
        if ulang != "y":
            print("\n Terima kasih sudah bermain! Sampai jumpa!")
            break

num_guess()
```
